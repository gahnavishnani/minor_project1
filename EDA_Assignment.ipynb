```python
df = df.drop(columns=["Unnamed: 0.1", "Unnamed: 0"])
```

```python
df.shape
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["price"], bins=20, kde=True)
plt.title("Distribution of Laptop Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df["price"])
plt.title("Box Plot of Laptop Prices")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y=df["brand"], order=df["brand"].value_counts().index)
plt.title("Number of Laptops by Brand")
plt.xlabel("Count")
plt.ylabel("Brand")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="spec_rating", y="price")
plt.title("Specification Rating vs Laptop Price")
plt.xlabel("Specification Rating")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=df, x="brand", y="price")
plt.xticks(rotation=90)
plt.title("Laptop Price Distribution by Brand")
plt.xlabel("Brand")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=['int64', 'float64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# Final Summary & Insights

### Key Insights

1. The laptop price distribution is positively skewed, with most laptops priced between ₹40,000 and ₹90,000 and only a few premium laptops priced much higher.

2. Laptop price generally increases with higher specification ratings. Brands such as Apple and Microsoft tend to have higher-priced laptops, while HP, Lenovo, and Dell provide laptops across a broader price range.

3. The correlation analysis shows that specification rating and display resolution have a moderate positive relationship with laptop price. Warranty has little influence on laptop pricing.

### Model Readiness

The dataset is clean and ready for machine learning because:
- There are no missing values.
- There are no duplicate rows.
- Unnecessary index columns were removed.
